## ถ้าเอาตารางไม่เหมือนกันมา concat กัน

```python
KKC4 = data_df[data_df['province_of_onset']=='ขอนแก่น'][['sex','nationality']]
KKC4.head()
```

## List(row oriented)

```python
data = [['Jones LLC',150,200,140], #ข้อมูลในตาราง
        ['Alpha Co',200,210,215],
        ['Blue Inc',50,90,95]]
header = ['account','Jan','Feb','Mar'] #หัวตาราง
```

```python
df2 = pd.DataFrame.from_records(data,columns=header)
df2
```

## Dictionary(column oriented)

```python
data_dict = {'account':['Jones LLC','Alpha Co','Blue Inc'],
             'Jan':[150,200,50],
             'Feb':[200,210,90],
             'Mar':[140,215,95]    
}
```

```python
df3 = pd.DataFrame.from_dict(data_dict)
df3
```

## List(column oriented) #ติดบัคอยู่

```python
data_item = [['account',['Jones LLC','Alpha Co','Blue Inc']],
             ['Jan',[150,200,50]],
             ['Feb',[200,210,90]],
             ['Mar',[140,215,95]]
             ]
```

```python
df4 = pd.DataFrame.from_items(data_item)
```
